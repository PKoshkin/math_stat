```python
import numpy.random as rand
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
theta = 10
N = 10**4

values = np.random.uniform(0, theta, N)
```

```python
# Считаем модули разностей тетта и оценок

# Считаем 2 * <X>
difference1 = []
for n in range(1, N + 1):
    difference1.append(abs(2 * np.sum(values[:n] / n) - theta))
    
# Считаем <X> + 1/2 * X_(N)
difference2 = []
for n in range(1, N + 1):
    difference2.append(abs(np.sum(values[:n] / n) + values[:n].max() / 2 - theta))
    
# Считаем (n + 1) * X_(1)
difference3 = []
for n in range(1, N + 1):
    difference3.append(abs((n + 1) * values[:n].min() - theta))
    
# Считаем X_(1) + X_(n)
difference4 = []
for n in range(1, N + 1):
    difference4.append(abs(values[:n].min() + values[:n].max() - theta))
    
# Считаем (n + 1) / n * X_(n)
difference5 = []
for n in range(1, N + 1):
    difference5.append(abs((n + 1) * values[:n].max() / n - theta))
```

```python
# Строим все графики
myFigure, myAxis = plt.subplots()

myAxis.plot(difference1[10**3:], label=r"$2\bar{X}$")
myAxis.plot(difference2[10**3:], label=r"$\bar{X} + \frac{X_{(1)}}{2}$")
myAxis.plot(difference3[10**3:], label=r"$(n + 1)\bar{X}$")
myAxis.plot(difference4[10**3:], label=r"$X_{(1)} + X_{(n)}$")
myAxis.plot(difference5[10**3:], label=r"$\frac{(n+1)}{n}X_{(1)}$")
myAxis.legend(loc="upper left")
myAxis.set_xlabel("n")
myAxis.set_ylabel(r"$|evaluation - \theta|$")
myAxis.set_title(r"All evaluations with $\theta = {}$".format(theta))
```

```python
# Исключаем хреновую оценку
myFigure, myAxis = plt.subplots()

myAxis.plot(difference1[10**3:], label=r"$2\bar{X}$")
myAxis.plot(difference2[10**3:], label=r"$\bar{X} + \frac{X_{(1)}}{2}$")
myAxis.plot(difference4[10**3:], label=r"$X_{(1)} + X_{(n)}$")
myAxis.plot(difference5[10**3:], label=r"$\frac{(n+1)}{n}X_{(1)}$")
myAxis.legend(loc="upper left")
myAxis.set_xlabel("n")
myAxis.set_ylabel(r"$|evaluation - \theta|$")
myAxis.set_title(r"All evaluations with $\theta = {}$".format("{(1)}", theta))
```

```python
# Исключаем еще оценок
# Остались лучшие оценки
myFigure, myAxis = plt.subplots()
myAxis.plot(difference4[10**3:], label=r"$X_{(1)} + X_{(n)}$")
myAxis.plot(difference5[10**3:], label=r"$\frac{(n+1)}{n}X_{(1)}$")
myAxis.legend(loc="upper left")
myAxis.set_xlabel("n")
myAxis.set_ylabel(r"$|evaluation - \theta|$")
myAxis.set_title(r"Best evaluations with $\theta = {}$".format("{(1)}", theta))
```
